```python
import pandas as pd

data = pd.read_clipboard() # 读取家族攻击力统计表
```

```python
data = data.drop(columns=['分队', '序号', '攻击力'])
```

```python
for i in range(1, 10):
    temp_data = data[data['所属'] == i].copy()
    name_ls = temp_data['昵称'].values

    group_info = ", ".join(name_ls)
    print(f"第{i}组: {group_info}")
    
```
